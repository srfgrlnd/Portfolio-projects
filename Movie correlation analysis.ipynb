## Correlation analysis on movie dataset
### Which variable affects the gross earnings of a movie?

```python
# import libraries

import numpy as np
import pandas as pd
import seaborn as sns

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # Adjusts configurations of plots

```

```python
# Read in the data


df = pd.read_csv(r'C:\Users\fager\OneDrive\Skrivbord\Portfolio projects\movies.csv')
```

```python
# Check data

df.head()
```

```python
# check missing data

df.isna().sum()
```

```python
# check the types and Non-Null counts again

df.info()
```

```python
# remove missing values

df = df.dropna()
```

```python
# missing values are dropped

df.isna().sum()
```

```python
# change data type for columns 'budget' and 'gross' from float64 to int64

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
# check that the data types have changed

df.dtypes
```

```python
# add new column 'yearcorrect' that includes the correct year since the 'released' and
# 'year' column in the dataset do not always match

df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)
```

```python
df.head()
```

```python
# Sort data by gross revenue 

df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# Remove duplicate values

df.drop_duplicates()
```

```python
# check Pearson correlation between values (only numericals)

# budget and gross have the highest correalation 0.74, votes and gross 0.61

df.corr()
```

```python
#correlation matrix

corr_matrix = df.corr()

sns.heatmap(corr_matrix, annot=True)

plt.title('Correlation matrix for numerical features')

plt.show()
```

```python
# Focus on budget and gross.
# scatterplot for budget and gross

plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budget vs Gross Earnings')

plt.xlabel('Budget for Film')
plt.ylabel('Gross Earnings')

plt.show()
```

```python
# visualize correlation with seaborn plot

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color": "red"}, line_kws={"color":"blue"})
```

```python
# Look at categorical values 
# need to convert categorical values to numerical values in ordet to get correlation

df_num = df

for col_name in df_num.columns:
    if(df_num[col_name].dtype == 'object'):
        df_num[col_name] = df_num[col_name].astype('category')
        df_num[col_name] = df_num[col_name].cat.codes
        
df_num.head()
```

```python
#correlation matrix

corr_matrix = df_num.corr()

sns.heatmap(corr_matrix, annot=True)

plt.title('Correlation matrix for numerical features')

plt.show()
```

```python
# Since I am interested in variables correlating with gross I will 
# get only the Gross correlation values

corr_mat_num = df_num.corr()

corr_pairs = corr_mat_num.unstack()

corr_pairs['gross'].sort_values(ascending=False)
```

### The correlation analysis shows that the variables Budget and Votes affect the gross earnings
